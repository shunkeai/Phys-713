From the conservation of energy and the equation of state of ideal gas, we have
\begin{eqnarray}
{v_1^2 \over 2}+\epsilon_1 + {P_1 \over \rho_1} = {v_0^2 \over 2}+\epsilon_0 + {P_0 \over \rho_0}
\end{eqnarray}

\begin{eqnarray}
{P \over \rho}=({\gamma-1})\epsilon
\end{eqnarray}
where $\gamma=5/3$.
\begin{eqnarray}
{v_0 \over v_1}={(\gamma+1)P_1 + (\gamma-1)P_0 \over (\gamma-1)P_1 + (\gamma+1)P_0}
\end{eqnarray}

For strong shock, $\epsilon_1 >> \epsilon_0$, $P_1 >> P_0$, 
\begin{eqnarray}
{v_0 \over v_1} = {(\gamma+1) + (\gamma-1)(P_0/P_1) \over (\gamma-1) + (\gamma+1)(P_0/P_1)}={\gamma+1 \over \gamma-1}
\end{eqnarray}

\begin{eqnarray}
{v_1^2 \over 2} + \gamma \epsilon_1 = {v_0^2 \over 2}+\gamma \epsilon_0 = {v_0^2 \over 2}
\end{eqnarray}

\begin{eqnarray}
{v_1^2 \over v_0^2}+{\gamma \epsilon \over (1/2)v_0^2}=1
\end{eqnarray}

\begin{eqnarray}
{\epsilon \over (1/2)v_0^2}=[1-({v_1 \over v_0})^2]{1 \over \gamma}={4 \over (\gamma+1)^2}={9 \over 16}
\end{eqnarray}
