(a).
\begin{eqnarray}
{\rm SNR_{1m}}={N_{\rm star,1m} \over \sqrt{N_{\rm star,1m}+N_{\rm sky,1m}}}
={N_{\rm star,1m} \over \sqrt{N_{\rm star,1m}+n_{\rm sky,1m}\pi R^2}}
\end{eqnarray}

(b).
\begin{eqnarray}
{N_{\rm star,10m} \over N_{\rm star,1m}}={(10m)^{2} \over (1m)^{2}}=100
\end{eqnarray}

\begin{eqnarray}
{n_{\rm sky,10m} \over n_{\rm sky,1m}}={(10m)^{2} \over (1m)^{2}}=100
\end{eqnarray}

\begin{eqnarray}
{N_{\rm sky,10m} \over N_{\rm sky,1m}}={n_{\rm sky,10m} \pi (R/3)^2 \over n_{\rm sky,1m} \pi R^2}={100 \over 9}
\end{eqnarray}

\begin{eqnarray}
{\rm SNR_{10m}}={N_{\rm star,10m} \over \sqrt{N_{\rm star,10m}+N_{\rm sky,10m}}}={100N_{\rm star,1m} \over \sqrt{100N_{\rm star,1m}+{100\over 9}N_{\rm sky,10m}}}=10 {N_{\rm star,1m} \over \sqrt{N_{\rm star,1m}+{1\over 9}N_{\rm sky,10m}}}=10 {N_{\rm star,1m} \over \sqrt{N_{\rm star,1m}+{1 \over 9}n_{\rm sky,1m}\pi R^2}}
\end{eqnarray}

(c).
Assume the accumulated rate of photons from the star is $\dot{N_{\rm sky}}$. For each pixel, assume the acculated rate of sky photons is $\dot{n}_{\rm sky}$ 
\begin{eqnarray}
{\rm SNR}={\dot{N}_{\rm star}t \over \sqrt{\dot{N}_{\rm star}t+\dot{n}_{\rm sky}t\pi R^2}}={\dot{N}_{\rm star} \over \sqrt{\dot{N}_{\rm star}+\dot{n}_{\rm sky}\pi R^2}}t^{1/2} \propto t^{1/2}
\end{eqnarray}

\begin{eqnarray}
{\rm SNR_{source-limited}}=\sqrt{\dot{N}_{\rm star}}t^{1/2}
\end{eqnarray}

\begin{eqnarray}
{\rm SNR_{background-limited}}={\dot{N}_{\rm star} \over \sqrt{\dot{n}_{\rm sky}\pi R^2}}t^{1/2}
\end{eqnarray}

(d).

In the "source-limited" case, 
\begin{eqnarray}
{\rm SNR_{10m}}=10\sqrt{\dot{N}_{star,1m}}~t^{1/2}=\sqrt{\dot{N}_{star,1m}}~(100t)^{1/2}
\end{eqnarray}

In the "background-limit" case,
\begin{eqnarray}
{\rm SNR_{10m}}=10\times 3 {\dot{N}_{\rm star, 1m} \over \sqrt{\dot{n}_{\rm sky, 1m}\pi R^2}}t^{1/2}\sim {\dot{N}_{\rm star, 1m} \over \sqrt{\dot{n}_{\rm sky, 1m}\pi R^2}}(1000t)^{1/2}
\end{eqnarray}

Hence, in "source-limited" case, the exposure time t of 1-m telescope needed to be increased with a factor $100$ to reach the S/N obtained with the 10-m telescope, while that factor should be $\sim 1000$ in the "background-limited" case.


# Chapter 2, question 1
(a).
Since the intensity (I) is a constant along any ray and isotropic in $\Omega$,

\begin{eqnarray}
\Omega=\int sin\theta d\theta d\phi=2\pi (1-cos\theta) 
\end{eqnarray}

\begin{eqnarray}
f(d_{\odot})= -I {cos^2 \theta \over 2}|^{\theta}_0\phi|^{2\pi}_0=-2\pi I ({cos^2 {\theta}-1 \over 2})=\pi I (1+cos\theta)(1-cos\theta)={1 \over 2} I \Omega (1+cos \theta)
\end{eqnarray}

Since $d_{\odot}>>R_{\odot}$, $cos \theta \sim 1$, $f(d_{\odot})= I \Omega$.

At the surface of the sun, 
\begin{eqnarray}
f(0)=\pi I =\pi f(d_{\odot})/ \Omega
\end{eqnarray}

I_{\lambda_1}={2hc^2 \over \lambda^5_1}{1 \over {\rm exp}(hc/\lambda_1 kT)-1}
\end{eqnarray}

\begin{eqnarray}
f_{\lambda_1}=I_{\lambda_1}\Omega={2hc^2 \over \lambda^5_1}{\Omega \over {\rm exp}(hc/\lambda_1 kT+1)}
\end{eqnarray}

\begin{eqnarray}
{\rm exp}(hc/\lambda_1 kT)={2hc^2 \over \lambda^5_1}{\Omega \over f_{\lambda}(\lambda_1)}+1
\end{eqnarray}

\begin{eqnarray}
T={hc \over \lambda_1 k}[ln({2hc^2 \over \lambda_1^5}{\Omega \over f_{\lambda}(\lambda_1)})+1]^{-1},
\end{eqnarray}

where $h$ is the Planck constan, and $k$ is the Boltzmann constant.
